```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from copy import deepcopy
```

```python
THIS_FOLDER = os.path.dirname(os.path.abspath(''))
print(THIS_FOLDER)
SAVED_WEIGHT = os.path.join(THIS_FOLDER, 'ver2_result')
print(SAVED_WEIGHT)
```

```python
scores = np.load(os.path.join(SAVED_WEIGHT, 'scores.npy'))
losses = np.load(os.path.join(SAVED_WEIGHT, 'J.npy'))
```

```python
scores_avg = deepcopy(scores)
losses_avg = deepcopy(losses)
step = 25
for i in range(step - 1, len(scores_avg), step):
    start = i + 1 - step
    end = i + step if (i + step < len(scores_avg)) else len(scores_avg)
    avg = np.average(scores[start:end])
    scores_avg[start:end] = [avg] * (end - start)

for i in range(step - 1, len(losses_avg), step):
    start = i + 1 - step
    end = i + step if (i + step < len(losses_avg)) else len(losses_avg)
    avg = np.average(losses[start:end])
    losses_avg[start:end] = [avg] * (end - start)
```

```python
plt.plot(scores_avg)
```

```python
plt.plot(losses_avg)
```
